```python
a = np.random.randint(1,99,size=(5,5))
d = np.diag(a) # get diagonal
print(a)
print(d) # get diagonal of a

# time complexity
import time
start_time = time.time()
a = np.random.randint(1,99,size=(1000,1000))
print(a)
print("--- %s seconds ---" % (time.time() - start_time))

#print diagonal time 
start_time = time.time()
print(np.diag(a))
print("--- %s seconds ---" % (time.time() - start_time))

```
